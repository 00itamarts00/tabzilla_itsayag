```python
# read cleaned metafeatures file
keep_metafeatures_df = pd.read_csv("./results/cleaned_metafeatures.csv", index_col=0)

# drop cols that are all NaN

# keep_metafeatures_df.loc[keep_metafeatures_df.isna().sum() > 0, :]
na_feats = keep_metafeatures_df.isna().sum() == len(keep_metafeatures_df)
na_feat_names = list(na_feats[na_feats].index)
print(f"dropping cols with only NA: {na_feat_names}")
keep_metafeatures_df.drop(columns=na_feat_names, inplace=True)

# replace inf with something else
keep_metafeatures_df.replace([np.inf, -np.inf], 0.0, inplace=True)

all_feature_cols = [c for c in keep_metafeatures_df.columns if c.startswith("f__")]

def simplify(x):
    return x.replace("statistical", "stat").replace("model-based", "model").replace("landmarking", "landmark").replace("skewness", "skew")

nice_feature_names = [simplify(f[len("f__pymfe."):]) for f in all_feature_cols]
```

## merge performance dataset with metafeatures

```python
df = metadataset_df.loc[metadataset_df["alg_name"].isin(keep_algs), :]

# df = df.merge(keep_metafeatures_df, left_on="dataset_fold_id", right_on="dataset_name", how="left")

# label each alg by their category
for name, category in keep_algs.items():
    df.loc[df["alg_name"] == name, "alg_category"] = category 
```

```python
df["alg_name"].value_counts()
```

```python
# re-assign ranks of each alg
# metric_col = "normalized_Accuracy__test"
metric_col = "normalized_Log Loss__test"

# first take best performance of each category
df_best_category = df.groupby(["dataset_fold_id", "alg_category"])[metric_col].min().reset_index()
# df_best_category = df.groupby(["dataset_fold_id", "alg_category"])[metric_col].max().reset_index()

df_best_category.loc[:, f"rank"] = df_best_category.groupby(["dataset_fold_id"])[metric_col].rank(method="min", ascending=True).values
# df_best_category.loc[:, f"rank"] = df_best_category.groupby(["dataset_fold_id"])[metric_col].rank(method="min", ascending=False).values
```

```python
df_best_category.head(10)
```

```python
# take only the top-ranked algs
df_top = df_best_category.loc[df_best_category["rank"] == 1, :]

# best_algs_category = df_top.groupby("dataset_fold_id")["alg_category"].apply(lambda x: x if len(x) == 1 else 5).reset_index()
best_algs = df_top.groupby("dataset_fold_id")["alg_category"].apply(lambda x: set(x)).reset_index()

best_algs.loc[:, "num_winners"] = best_algs["alg_category"].apply(lambda x: len(x))

best_algs.loc[:, "winning_category"] = 0
best_algs.loc[best_algs["num_winners"] == 1, "winning_category"] = best_algs.loc[best_algs["num_winners"] == 1, "alg_category"].apply(lambda x: list(x)[0])
best_algs.loc[best_algs["num_winners"] > 1, "winning_category"] = "multiple"

print(len(best_algs))
print(best_algs["alg_category"].value_counts())
print(best_algs["winning_category"].value_counts())

best_algs.loc[:, "alg_category_str"] = best_algs["alg_category"].astype(str)
```

```python
# merge dataset features
df = best_algs.merge(keep_metafeatures_df, left_on="dataset_fold_id", right_on="dataset_name", how="left")

# make a version with only one winner
df_one_winner = df.loc[df["winning_category"] != "multiple", :]
```

```python
from sklearn import tree, metrics

# keep_feature_cols = [c for c in df.columns if c.startswith("f__pymfe.")]  # c.startswith("f__pymfe.info") or c.startswith("f__pymfe.complexity") or c.startswith("f__pymfe.statistical") or c.startswith("f__pymfe.clustering") or c.startswith("f__pymfe.general")]
keep_feature_cols = [c for c in df.columns if (c.startswith("f__pymfe.info") or c.startswith("f__pymfe.complexity") or c.startswith("f__pymfe.statistical") or c.startswith("f__pymfe.clustering") or c.startswith("f__pymfe.general")) and ("hist" not in c)]

nice_feature_names = [simplify(f[len("f__pymfe."):]) for f in keep_feature_cols]

x = df_one_winner[keep_feature_cols].values.clip(-1e10, 1e10)
y = df_one_winner["winning_category"] # df["alg_category_str"]  #

# x = df[keep_feature_cols].values.clip(-1e10, 1e10)
# y =  df["winning_category"] # df["alg_category_str"]  #

# train a single tree of max-depth 5 to visualize
clf_display = tree.DecisionTreeClassifier(max_depth=3, min_samples_leaf=20, random_state=9999).fit(x, y)

print(clf_display.classes_)

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(20, 8))

annotations = tree.plot_tree(
    clf_display, 
    # max_depth=2, 
    feature_names=nice_feature_names, 
    class_names=clf_display.classes_,
    filled=True, 
    ax=ax,
    fontsize=8,
    rounded=True,
    node_ids=True,
)
# print(tree.export_text(clf_display, feature_names=feature_cols))
plt.show()
print(f"using {len(x)} instances of 1705")
print(f"accuracy for this model: {metrics.accuracy_score(y, clf_display.predict(x))}")
```

```python
# PLOT_LABELS = ["GBDT", "NODE", "ResNet", "SAINT", "TabPFN"]
# COLORS = ["darkgreen", "firebrick", "darkviolet", "darkorange", "skyblue"]
PLOT_LABELS = ["XGBoost", "CatBoost", "SAINT", "ResNet", "TabPFN"]
COLORS = ["teal", "forestgreen", "goldenrod", "darkviolet", "royalblue"]
def plot_bar(data_dict, ax):

    # fig, ax = plt.subplots(figsize=(2.5, 2))
    data = []
    for l in PLOT_LABELS:
        if l in data_dict:
            data.append(data_dict[l])
        else:
            data.append(0)
    y_pos = np.arange(len(data))

    # only color the winning alg
    lose_color = "lightgray"
    plot_colors = [lose_color for _ in data]
    winning_idx = np.argmax(data)
    plot_colors[winning_idx] = COLORS[winning_idx]

    hbars = ax.barh(y_pos, data, color=plot_colors)

    ax.set_title(PLOT_LABELS[winning_idx], color=COLORS[winning_idx], fontweight="bold")

    barlabels = ax.bar_label(hbars, fmt='%d', label_type="edge")

    for i, d in enumerate(data):
        if d == max(data):
            barlabels[i].set_font_properties({'weight': "bold"})
        if d < 5:
            ax.text(x=9, 
                    y=y_pos[i] - 0.1, 
                    s=f"{d:3d}",
                    horizontalalignment='right',
                    )
        # if d > 500:
            # barlabels[i].set_label_type({'size': 30})

    # # access Daniel's label and change properties
    # barlabels[-1].set_font_properties({'size': 30})
    # barlabels[-1].set_color('#E34234')

    # ax.set_xlim(right=15)
    ax.set_yticks(y_pos, labels=PLOT_LABELS)

    # plt.show()
```
